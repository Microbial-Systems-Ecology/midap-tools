You want to investigate bleaching in your experiment. your hypothesis is, that after cells stop growing (around frame 20) living cells get bleached more and more
You are interested in the bleaching rate of CFP and YFP positive cells to see if the cell types or fluorophores have different susceptabilities to bleaching.

Your task is to use the growth rate function in conjuction with `mean_intensity` to determine over time, how bleaching progresses

-   first, calcualte the growth rate based on the log transformed intensity_mean with a integration window of 10. call the resulting column `bleach_rate`
-   create a rate plot of the `bleach_rate`. create it for data aggregated within the entire experiment
-   for visualization, also create a rate plot of the `intensity_mean` and one or the `intensity_mean_log`.

```python
from fluid_experiment.fluid_experiment import FluidExperiment
PATH = "../../../data/midap-tools_example"
experiment = FluidExperiment(PATH,name = "example_experiment")
experiment.load_metadata_template()
print(experiment)

### YOUR SOLUTION BELOW

experiment.calculate_transform_data("intensity_mean")  # we log transform, so the intensity is normalized
experiment.calculate_growth_rate(integration_window=10,id_column="trackID",value_column="intensity_mean_log",growth_rate_column="bleach_rate")

experiment.plot_rates(rate_column= "bleach_rate", group_by="experiment", title="Bleach rate over time")
experiment.plot_rates(rate_column= "intensity_mean_log", group_by="experiment",title="Log intensity over time")
experiment.plot_rates(rate_column= "intensity_mean", group_by="experiment", title="Intensity over time")
```
